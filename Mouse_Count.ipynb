```python

# You have to use the encoding as latin1 to read this file as there are some special character in this file...
path = r'C:\Users\shossein\Downloads\MousCount'
save_path = r'C:\Users\shossein\Downloads\MousCount\MousCount' # DEFINE the file directory for saving

df = pd.read_csv(path + r"\Narcotic Administration - Pentobarbital.csv")  
df.head(20)
```

```python
import pandas as pd

data_category_range = df['Responsible\nPerson'].unique()
data_category_range = data_category_range.tolist()
data_category_range = [x for x in data_category_range if str(x) != 'nan']


for i,value in enumerate(data_category_range):
    df[df['Responsible\nPerson'] == value].to_csv(save_path + str(value)+ '.csv',index = False, na_rep = 'N/A')
```

```python
df['Mouse_ID'] = df['Mouse ID'].str[:3]
```

```python
df['Dates'] = df['Date'].str[-2:]
```

```python

for i,value in enumerate(data_category_range):

        df[df['Responsible\nPerson'] == value].to_csv(save_path +str(value)+r'.csv',index = False, na_rep = 'N/A')
        dff = pd.read_csv (save_path + str(value)+r'.csv')
        fig = px.sunburst(dff,  path=['Mouse_ID', 'Dates'], color='Amount\nadministered [ml]',  title= 'Researcher:  ' + str(value))
        fig.write_html(os.path.join(save_path  + str(value)+ "sunburst.html"))
        fig.show()
```

```python
for i,value in enumerate(data_category_range):

        df[df['Responsible\nPerson'] == value].to_csv(save_path +str(value)+r'.csv',index = False, na_rep = 'N/A')
        dff = pd.read_csv (save_path + str(value)+r'.csv')
        fig = px.treemap(dff,  path=[px.Constant('Mouselines'),'Mouse_ID', 'Dates'], color='Amount\nadministered [ml]',  title= 'Researcher:  ' + str(value))
        fig.write_html(os.path.join(path  + str(value)+ "treemap.html"))
        fig.show()
```

```python
for i,value in enumerate(data_category_range):

        df[df['Responsible\nPerson'] == value].to_csv(save_path +str(value)+r'.csv',index = False, na_rep = 'N/A')
        dff = pd.read_csv (save_path + str(value)+r'.csv')
        fig = px.bar(dff,  x= 'Mouse_ID', color="Dates",  title= 'Researcher:  ' + str(value))
        fig.write_html(os.path.join(save_path  + str(value)+ "bar.html"))
        fig.show()
        
```

```python

writer = pd.ExcelWriter(os.path.join(path,  'Mouse_count.xlsx'))
for i,value in enumerate(data_category_range):

  df[df['Responsible\nPerson'] == value].to_csv(save_path +str(value)+r'.csv',index = False, na_rep = 'N/A')
  dff = pd.read_csv (save_path + str(value)+r'.csv')

  dfff = pd.DataFrame(dff)
  

  print(tabulate(dfff, headers=['Mouseline ' + str(value), 'Count'], tablefmt='psql'))

  
  pd.DataFrame(dfff).to_excel(writer,sheet_name = str(value).format(i) , startcol=0)

                                          
writer.save()
```
